```python
import pandas as pd
import os

film_data_folder = "film-data"

df_list = []
# Loops over all csv files in the film data folder
for root, dir, files in os.walk(film_data_folder):
    for file in files:

        print(file)

        # Adds the csv file to a data frame and adds that data frame to a list of data frames
        df_list.append(pd.read_csv(os.path.join(root, file)))

# Combines all the data frames into one
df_films = pd.concat(df_list)
df_films
```

```python
df_films["Year"].unique()
```

```python
# Rename index for each year to popularity score (order they were taken from letterboxd is order they were ranked in popularity)

df_films = df_films.rename(columns={"Unnamed: 0": "Popularity in Year"}).reset_index(drop=True)
df_films["Popularity in Year"] = df_films["Popularity in Year"].apply(lambda x: x + 1)
df_films
```

```python
import ast
import re
import numpy as np

# Expand ratings text to columns for each star rating, and aggregate to get Num Ratings and Film Rating column

rating_nums = np.arange(0.5, 5.5, 0.5)

for index, row in df_films.iterrows():
    ratings = ast.literal_eval(row["Letterboxd Ratings"])
    ratings = {float(key): value for key, value in ratings.items()}
    for n in rating_nums:
        match = re.search(r'\b([\d,]+)\b', ratings[n])
        if(match != None):
            rating_value = int(match.group(0).replace(',', ''))
        else:
            rating_value = 0
        df_films.loc[index, f'{n}'] = rating_value

df_films["Num Ratings"] = df_films.loc[:, [str(n) for n in rating_nums]].sum(axis=1)
df_films["Film Rating"] = df_films.loc[:, [str(n) for n in rating_nums]].mul(rating_nums).sum(axis=1) / df_films["Num Ratings"]
df_films = df_films.drop(columns="Letterboxd Ratings")

df_films.head()

    
```

```python
import math

# Get standard deviation of rating. Useful for seeing if ratings are more spread out or more skewed towards one value. 

for index, row in df_films.iterrows():
    sd = math.sqrt(sum([math.pow((n - row["Film Rating"]), 2)*row[str(n)] for n in rating_nums])/row["Num Ratings"])
    df_films.loc[index, "SD Rating"] = sd

df_films
```

```python
# Convert Genres and Directors columns from string to list 
# Also remove any comma characters

df_films["Genres"] = df_films["Genres"].apply(lambda x: ast.literal_eval(x))
df_films["Directors"] = df_films["Directors"].apply(lambda x: ast.literal_eval(x))
df_films["Directors"] = df_films["Directors"].apply(lambda x: [[string.replace(",", "") if string else string for string in y] for y in x])
df_films.head()
```

```python
# Iterate through film directors to get a Directors data frame of unique director rows

all_directors = []

for index, row in df_films.iterrows():
    for director in row["Directors"]:
        all_directors.append(director)

df_directors = pd.DataFrame(all_directors, columns=["Name", "Gender", "Birthdate"])
df_directors = df_directors.drop_duplicates()
df_directors
```

```python
df_directors[df_directors["Name"].apply(lambda x: "Arthur" in x)]
```

```python
# String excess text off Gender column in Directors data frame

df_directors["Gender"] = df_directors["Gender"].apply(lambda x: x.lstrip("Gender\n") if x is not None else x)
df_directors["Gender"] = df_directors["Gender"].replace('-', None)
df_directors["Gender"].unique()
```

```python
# String excess text off Birthdate column in Directors data frame

df_directors["Birthdate"] = df_directors["Birthdate"].apply(lambda x: x.lstrip("Birthday\n") if x is not None else x)
df_directors["Birthdate"] = df_directors["Birthdate"].apply(lambda x: x.split("(")[0].strip() if x is not None else x)
df_directors["Birthdate"] = df_directors["Birthdate"].replace('-', None)

df_directors
                                                        
```

```python
from datetime import datetime

# Checking if all director birthdate are valid date format

def is_valid_date_format(date_string):
    if date_string==None: return False
    try:
        datetime.strptime(date_string, '%B %d, %Y')
        return True
    except ValueError:
        return False

df_directors.loc[df_directors["Birthdate"].apply(lambda x: not is_valid_date_format(x)), "Birthdate"].unique()
```

```python
df_directors["Birthdate"] = df_directors["Birthdate"].replace('Adult Actor\nTrue', None)
```

```python
# Now converting birthdates to datetime objects

df_directors["Birthdate"] = df_directors["Birthdate"].apply(lambda x: datetime.strptime(x, '%B %d %Y') if x is not None else None)

df_directors
```

```python
# Check how mnay duplicated director names. 
# Sometimes gender and birthdate information failed to be collected for a director for one film, 
# but was succesfully collected for another, creating different rows with duplicate director names. 

df_directors[df_directors.duplicated(subset='Name', keep=False)].sort_values(by="Name")
```

```python
# Removing rows that have a director name but no gender info, 
# only if there is another row with the same director name

for index, row in df_directors.iterrows():
    if row["Gender"] == None:
        if (df_directors["Name"] == row["Name"]).sum() > 1:
            df_directors = df_directors.drop(index)   

df_directors = df_directors.reset_index(drop=True)
```

```python
# Stripping director column in films data frame back to just a list of director names. 
# Gender and birthdate info can be found in Directors data frame now. 

df_films["Directors"] = df_films["Directors"].apply(lambda x: [x[n][0] for n in range(len(x))])
df_films
```

```python
## Adding genre column for directors df

director_genres_list = []

for index, row in df_directors.iterrows():
    df_films_filtered = df_films[df_films["Directors"].apply(lambda x: row["Name"] in x)]
    genres_list = [genre for genre_list in df_films_filtered["Genres"] for genre in genre_list]
    genres_list = list(set(genres_list)) ## getting unique items
    director_genres_list.append(genres_list)

df_directors["Genres"] = director_genres_list
df_directors
```

```python
# Adding a director Gender column in Films data frame containing list of Director genders pullled from Directors data frame

df_films["Directors Genders"] = df_films["Directors"].apply(lambda x: [df_directors.loc[df_directors["Name"]==name, "Gender"].iloc[0] for name in x])
df_films
```

```python
# Adding a director Age (at film release year )column in Films data frame containing list of Director genders pullled from Directors data frame

def get_ages(film_year, directors):
    return [film_year - df_directors.loc[df_directors["Name"]==x, "Birthdate"].iloc[0].year for x in directors]

df_films["Directors Ages"] = df_films.apply(
    lambda row: 
        get_ages(row["Year"], row["Directors"]),
        axis=1
)
df_films
```

```python
# Creating 3 boolean columns for has_(gender)_director in Films data frame for ease of access 

df_films["has_male_director"] = df_films["Directors Genders"].apply(lambda x: "Male" in x)
df_films["has_female_director"] = df_films["Directors Genders"].apply(lambda x: "Female" in x)
df_films["has_nonbinary_director"] = df_films["Directors Genders"].apply(lambda x: "Non-binary" in x)
df_films
```

```python
## Checking all genres names

genre_list = [genre for genre_list in df_films["Genres"] for genre in genre_list]
list(set(genre_list))
```

```python
# Cleaning up genre names

def fix_genre_names(genre_list):
    genre_list = ['Fantasy' 
                  if item == 'Fantasy Adventure, Heroism, And Swordplay' 
                  else item 
                  for item in genre_list]
    
    genre_list = ['Science Fiction' 
                  if (item == 'Imaginative Space Odysseys And Alien Encounters') 
                  | (item == 'Action-Packed Space And Alien Sagas')
                  | (item == 'Monsters, Aliens, Sci-Fi And The Apocalypse')
                  else item 
                  for item in genre_list]
    
    genre_list = ['Action' 
                  if (item == 'Superheroes In Action-Packed Battles With Villains') 
                  | (item == 'Intense Combat And Martial Arts')
                  else item 
                  for item in genre_list]
    
    
    genre_list = [item for item in genre_list 
                  if (item != 'Show All…') 
                  & (item != 'Epic Heroes')]
    
    return list(set(genre_list)) # making sure unique list
    
df_films["Genres"] = df_films["Genres"].apply(lambda x: fix_genre_names(x))
```
